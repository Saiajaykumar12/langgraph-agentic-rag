```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a:int, b:int):
    """
        It will return the sum of two numbers

        Args:
            a : First number
            b : Second number
    """
    return a + b

@tool 
def multiply_numbers(a:int, b:int):
    """
        It will retuen the product of two numbers

        Args:
            a : First number
            b : Second number
    """
    return a * b
```

```python
from langchain_groq import ChatGroq
from langchain.agents import create_agent
llm = ChatGroq(model="llama-3.3-70b-versatile")
```

```python
agents = create_agent(
    model = llm,
    tools = [add_numbers, multiply_numbers],
    system_prompt = "You are a maths teacher, and always use tool for action."
)
```

```python
response = agents.invoke({"messages":[{"role":"user","content":"What is the sum of two numbers"}]})
```
